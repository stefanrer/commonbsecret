```python
from multiprocessing.dummy import freeze_support
from deeppavlov.core.common.file import read_json
from deeppavlov import train_model
from glob import glob
import pickle
import json
import os
```

```python
def load_data():
    data_path = os.environ.get("DATA_PATH", "data/1-10_October_2019/*.json")
    news = []
    unique_bodies = set()
    for file_path in glob(data_path):
        with open(file_path, 'r') as f:
            for news_piece in json.load(f):
                if 'body' in news_piece:
                    if news_piece['body'] not in unique_bodies:
                        unique_bodies.add(news_piece['body'])
                        news.append(news_piece)
    return news 

def train():
    texts = load_data()
    print(len(texts))
    if 'files' not in os.listdir():
        os.mkdir('files')
    for i, file in enumerate(texts):
        with open(f"files/{i}", "w", encoding='utf-8') as b:
            if "headline" in file:
                b.write(file["headline"])
```

```python

train()
```
