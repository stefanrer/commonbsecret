```python
import json
import pandas as pd
from collections import defaultdict
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
def get_link(u, link_tos):
    for k, v in link_tos.items():
        for link in v:
            if link.lower() in u['text'].lower():
                return k
    return None


def get_links_usage(data, link_tos):
    total_links = defaultdict(int)
    success_links = defaultdict(int)
    total_links_ratings = defaultdict(list)
    success_links_ratings = defaultdict(list)
    for dialog in data:
        rating = dialog['rating']
        utts = dialog['utterances']
        total_links_rating = {}
        success_links_rating = {}
        for i, u in enumerate(utts):
            if u.get('active_skill'):
                link = get_link(u, link_tos)
                if link:
                    total_links[link] += 1
                    if total_links_rating.get(link) is None:
                        total_links_rating[link] = rating
                    if i < len(utts) - 2:
                        if utts[i + 2].get('active_skill') == link:
                            success_links[link] += 1
                            if success_links_rating.get(link) is None:
                                success_links_rating[link] = rating
        for k in total_links_rating.keys():
            total_links_ratings[k].append(total_links_rating[k])
        for k in success_links_rating.keys():
            success_links_ratings[k].append(success_links_rating[k])
            
    percentage_links = {}
    total_links_percentage = {}
    total_links_count = sum([total_links[k] for k in total_links.keys()])
    for k in total_links.keys():
        percentage_links[k] = success_links[k] / total_links[k]
        total_links_ratings[k] = sum(total_links_ratings[k]) / len(total_links_ratings[k])
        success_links_ratings[k] = sum(success_links_ratings[k]) / len(success_links_ratings[k])
        total_links_percentage[k] = total_links[k] / total_links_count
    keys = list(sorted(total_links.keys()))
    def get_v(l):
        return [l[k] for k in keys]
        
    return {'total_links': get_v(total_links), 'success_links': get_v(success_links), 'percentage_links': get_v(percentage_links), 
            'total_links_ratings': get_v(total_links_ratings), 'success_links_ratings': get_v(success_links_ratings),
            'keys': keys, 'total_links_percentage': get_v(total_links_percentage)}
```

```python
def append_plots(html_base, new_html):
    if html_base == None:
        return new_html
    plot_to_add = new_html.split('\n')[3:-2]
    html_base = '\n'.join(html_base.split('\n')[0:-2] + plot_to_add + html_base.split('\n')[-2:])
    return html_base
merged_html = ''
for p in all_plots:
    merged_html = append_plots(merged_html, p.to_html())
with open(f'./static/plot.html', 'w') as fout:
        fout.write(merged_html)
```
